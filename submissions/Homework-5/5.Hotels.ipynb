````python
from bs4 import BeautifulSoup
import argparse
import requests
import logging
import pickle
import time
import sys
import os


log = logging.getLogger(__name__)
log.setLevel(logging.INFO)
loghandler = logging.StreamHandler(sys.stderr)
loghandler.setFormatter(logging.Formatter("[%(asctime)s] %(message)s"))
log.addHandler(loghandler)

base_url = "http://www.tripadvisor.com/"
user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2272.76 Safari/537.36"

def get_city_page(city, state, datadir):
    # Build the request URL
    url = base_url + "city=" + city + "&state=" + state
    
    # Request the HTML page
    headers = {'User-Agent': user_agent}
    response = requests.get(url, headers=headers)
    html = response.text.encode('utf-8')
    
    with open(os.path.join(datadir, city + '-tourism-page.html'), "w") as h:
        h.write(html)
    
    # get list of hotels
    soup = BeautifulSoup(html)
    li = soup.find("li", {"class": "hotels twoLines"})
    city_url = li.find('a', href=True)
    return city_url['href']


def get_hotellist_page(city_url, page_count, city, datadir='data/'):
    #get url
    url = base_url + city_url
    
    # Request page
    headers = { 'User-Agent' : user_agent }
    response = requests.get(url, headers=headers)
    html = response.text.encode('utf-8')
    # Save the webpage
    with open(os.path.join(datadir, city + '-hotelist-' + str(page_count) + '.html'), "w") as h:
        h.write(html)
    
    # find button to next page
    soup = BeautifulSoup(html)
    div = soup.find('a', {'class' : 'guiArw sprite-pageNext '})
    # button does not exist => last page
    if div is None:
        next_url = None
    # get url of next page from button
    else:
        next_url = div['href']
    
    return html, next_url


def parse_hotellist_page(html):
    # get html soup
    soup = BeautifulSoup(html)
    
    urls = []
    # get hotel urls from page
    hotels = soup.findAll('a', {'class': 'property_title'})
    for h in hotels:
        urls.append(h['href'])
        
    return urls
    

def scrape(city, state, data_dir='data/'):
    # get currentwd
    current_dir = os.getcwd()
    if not os.path.exists(os.path.join(current_dir, data_dir)):
        os.makedirs(os.path.join(current_dir, data_dir))
    
    # set of urls to hotel pages
    hotel_urls = set()
    # get the url for the hotel list page
    next_city_url = get_city_page(city, state, data_dir)
    c = 0
    while(True):
        c += 1
        # get html, and the url for the next page
        html, next_city_url = get_hotellist_page(next_city_url, c, city, data_dir)
        # get the list of hotel urls on the page
        hotel_urls.update(parse_hotellist_page(html))
        
        # Print number of hotels currently, and current page
        print len(hotel_urls), c
        # return if there isn't a next hotel url
        if next_city_url is None:
            break
        
        # sleep before requesting new page
        time.sleep(1)
    
    # save the urls to the hotel pages
    with open('hotelurls.pickle', 'wb') as f:
        pickle.dump(hotel_urls, f)
        print 'Done Writing'
````

````python
def parse_hotel(hotel_url):
    # Get url of hotel page
    url = base_url + hotel_url
    
    # Get HTML soup
    headers = { 'User-Agent' : user_agent }
    response = requests.get(url, headers=headers)
    html = response.text.encode('utf-8')
    
    soup = BeautifulSoup(html)
    
    
    # name of hotel
    hotelname = str(soup.find('h1', {'class' : 'header'}).string)
    
    # Get Ratings
    ratings = []
    ratings_rows = soup.find('ul', {'class' : 'barChart'})
    for r in ratings_rows.findAll('div', {'class' : 'wrap row'}):
        ratings.append(int(r.find('span', {'class' : 'compositeCount'}).string.replace(',', '')))
    
    # Get types of consumers
    consumers = []
    for c in soup.findAll('div', {'class' : 'segment'}):
        consumers.append(int(c.find('div', {'class' : 'value'}).string.replace(',', '')))

    # Get Ratings for qualities
    qualities = []
    div = soup.find('div', {'id' : 'SUMMARYBOX'})
    for c in div.findAll('img', {'class' : 'sprite-rating_s_fill'}):
        qualities.append(float(c['alt'].split()[0]))
    
    # Calculate avg rating
    total_ratings = float(sum(ratings))
    avg_rating = (ratings[0] * 5 + ratings[1] * 4 + ratings[2] * 3 + ratings[3] * 2 + ratings[4] * 1)/total_ratings
    
    # Whether the hotel is excellent or not
    excellent = int(ratings[0]/total_ratings >= .6)
    
    # Data vector
    vector = [hotelname] + [excellent] + [avg_rating] + ratings + consumers + qualities
    return vector
        
    
````

````python
# Load hotel urls
with open('hotelurls.pickle', 'rb') as f:
    hotel_urls = pickle.load(f)
    print(len(hotel_urls))

# get data
hotel_data = []
for h in hotel_urls:
    h_data = parse_hotel(h)
    hotel_data.append(h_data)
    time.sleep(.5)

# save hotel data
with open('hoteldata.pickle', 'wb') as f:
    pickle.dump(hotel_data, f)
````

````python
import numpy as np
import random

# Load data scraped from hotels
with open('hoteldata.pickle', 'rb') as f:
    hotel_data = pickle.load(f)
# Labels for a data row
data_descriptions = ['name', 'excellence', 'avg ratings', '5 rating', '4 rating', '3 rating', '2 rating', '1 rating', 'families',
                     'couples', 'solo', 'business', 'location', 'sleep quality', 'rooms', 'service', 'value', 'cleanliness']

# Number of hotels scraped
print 'Numer of hotels: ' + str(len(hotel_data))


# Shuffle data, and create train and test set
random.shuffle(hotel_data)
train_samples = int(round(len(hotel_data) * .8))

Xtrain = [x[8:] for x in hotel_data[:train_samples]]
ytrain = [x[2] for x in hotel_data[:train_samples]]

Xtest = [x[8:] for x in hotel_data[train_samples:]]
ytest = [x[2] for x in hotel_data[train_samples:]]


# Least Squares on training data
lstsq_data = np.linalg.lstsq(Xtrain, ytrain)
weights = lstsq_data[0]

# Calculate mean squared error for xdata and ydata
def mse(Xdata, ydata, weights):
    totalse = 0
    for x, y in zip(Xdata, ydata):
        pred = np.dot(x, weights)
        diff = (pred - y)**2
        totalse += diff
    mse = totalse/len(ydata)
    return mse

# MSE for Train and Test set
print 'Mean Squared Error, Train : ' + str(mse(Xtrain, ytrain, weights))
print 'Mean Squared Error, Test  : ' + str(mse(Xtest, ytest, weights))

# Coefficients from the Linear Regression, and their corresponding variable
print '\nCoefficients for Regression'
for d, w in zip(data_descriptions[8:], weights):
    if w >= 0:
        print d.ljust(14) + ':  ' + str(w)
    else:
        print d.ljust(14) + ': ' + str(w)
````

We can see from this that families, solo people, and business people have negative coefficients, which means that the more of them you have, the more likely it is that your hotel is going to have a lower rating. This seems to indicate that they rate the hotel harshly relative to the other groups of people.

This is compared to couples which have a positive coefficient, which indicates that couples would have a positive influence on your rating since they tend to rate hotels higher.

For the rest of the properties, such as value, and location, there is a positive influence between their ratings and the avg rating of the hotel, which matches what we would expect to see.

````python
from sklearn.linear_model import LogisticRegression
import random

# Load data scraped from hotels
with open('hoteldata.pickle', 'rb') as f:
    hotel_data = pickle.load(f)
# Labels for a data row
data_descriptions = ['name', 'excellence', 'avg ratings', '5 rating', '4 rating', '3 rating', '2 rating', '1 rating', 'families',
                     'couples', 'solo', 'business', 'location', 'sleep quality', 'rooms', 'service', 'value', 'cleanliness']

# Number of hotels scraped
print 'Numer of hotels: ' + str(len(hotel_data))


# Shuffle data, and create train and test set
random.shuffle(hotel_data)
train_samples = int(round(len(hotel_data) * .8))

Xtrain = np.array([x[8:] for x in hotel_data[:train_samples]])
ytrain = np.array([x[1] for x in hotel_data[:train_samples]])

Xtest = np.array([x[8:] for x in hotel_data[train_samples:]])
ytest = np.array([x[1] for x in hotel_data[train_samples:]])


# Logistic Regression Model
model = LogisticRegression()
model = model.fit(Xtrain, ytrain)

# Train and test scores
score_train = model.score(Xtrain, ytrain)
score_test  = model.score(Xtest, ytest)

print
print 'Training Score:', score_train
print 'Testing Score :', score_test

# coefficients of logit
coefs = model.coef_.T
print '\nCoefficients'
for label, c in zip(data_descriptions[8:], coefs):
    c = c[0]
    if c >= 0:
        print label.ljust(14) + ':  ', c
    else:
        print label.ljust(14) + ': ', c
````
